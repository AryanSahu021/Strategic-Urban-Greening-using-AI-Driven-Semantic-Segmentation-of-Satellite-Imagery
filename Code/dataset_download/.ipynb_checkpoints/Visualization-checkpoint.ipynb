```python
import ee
import folium

ee.Initialize()

latitude = 33.4484
longitude = -112.0740

buffer_radius = 0.1

aoi = ee.Geometry.Polygon([
    [
        [longitude - buffer_radius, latitude - buffer_radius],
        [longitude - buffer_radius, latitude + buffer_radius],
        [longitude + buffer_radius, latitude + buffer_radius],
        [longitude + buffer_radius, latitude - buffer_radius],
        [longitude - buffer_radius, latitude - buffer_radius]
    ]
])

startDate = '2022-04-01'
endDate = '2022-10-31'

image = ee.ImageCollection('LANDSAT/LC09/C02/T1_L2') \
    .filterDate(startDate, endDate) \
    .filterBounds(aoi) \
    .median()

lst = image.select('ST_B10').multiply(0.00341802).subtract(273.15-149)

lst_vis = {
    'min': 15,
    'max': 50,
    'palette': ['040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
               '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
               '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
               'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
               'ff0000', 'de0101', 'c21301', 'a71001', '911003']
}

map_center = aoi.centroid().coordinates().reverse().getInfo()

m = folium.Map(location=map_center, zoom_start=10)

folium.TileLayer(
    tiles=lst.getMapId(lst_vis)['tile_fetcher'].url_format,
    attr='Google Earth Engine',
    overlay=True,
    name='LST AOI',
    opacity=0.5
).add_to(m)

m
```
